```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
soccer_file = 'soccer_database/fifa_ranking.csv'
soccer_df = pd.read_csv(soccer_file)
soccer_df.head()
```

```python
world_happiness_file = 'world_happiness/2019.csv'
world_happiness_df = pd.read_csv(world_happiness_file)
world_happiness_df.head()
```

```python
# Transform soccer dataframe
soccer_cols = ['rank', 'country_full']
soccer_transformed = soccer_df[soccer_cols].copy()
soccer_transformed = soccer_transformed.rename(columns={'rank': 'FIFA Rank',
                                                        'country_full': 'Country'})
soccer_transformed = soccer_transformed.drop_duplicates(subset=['Country'])
soccer_transformed.head()
```

```python
# Transform world happiness dataframe
world_happiness_cols = ['Overall rank', 'Country or region', 'Score', 'GDP per capita',
                        'Healthy life expectancy', 'Freedom to make life choices',
                       'Perceptions of corruption']
world_happiness_transformed = world_happiness_df[world_happiness_cols].copy()
world_happiness_transformed = world_happiness_transformed.rename(columns={'Overall rank': 'Happiness Rank'})
world_happiness_transformed = world_happiness_transformed.drop_duplicates(subset=['Country or region'])
world_happiness_transformed.head()
```

```python
connection_string = "postgres:palooza@localhost:5432/soccer_and_happiness"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
soccer_transformed.to_csv('soccer_rank.csv', index=False)
world_happiness_transformed.to_csv('world_happiness.csv', index=False)
```
